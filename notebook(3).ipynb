Import Libraries

Start by importing the required libraries.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
```

Create the DataFrame

Create a DataFrame with the historical data.
Snice we are using a very small data set lets use a data frame 

```python
data = {
    "Year": [2009, 2011, 2012, 2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024],
    "Price (GH¢)": [8.00, 9.00, 11.00, 12.00, 14.00, 16.00, 18.00, 24.00, 36.00, 54.00, 75.00, 85.00, 95.00, 105.00]
}

df = pd.DataFrame(data)
```

 Visualize the Data

Plot the historical data to understand the trend.

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Price (GH¢)'], marker='o', linestyle='-', color='b')
plt.title('Average Price of Cement in Ghana (32.5R) 2009-2024')
plt.xlabel('Year')
plt.ylabel('Price (GH¢)')
plt.grid(True)
plt.show()
```

Train the Linear Regression Model

Prepare the data and train the model.

```python
# Prepare the data
X = df[['Year']]
y = df['Price (GH¢)']

# Train the linear regression model
model = LinearRegression()
model.fit(X, y)
```

 Make Predictions

Predict future prices.

```python
# Predict future prices
future_years = np.array([[2025], [2026], [2027], [2028], [2029]])
predicted_prices = model.predict(future_years)

# Create a DataFrame for the predictions
predictions_df = pd.DataFrame({
    'Year': future_years.flatten(),
    'Predicted Price (GH¢)': predicted_prices
})

print(predictions_df)
```

Visualize the Predictions

Plot the historical data along with the predictions.

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Price (GH¢)'], marker='o', linestyle='-', color='b', label='Historical Prices')
plt.plot(predictions_df['Year'], predictions_df['Predicted Price (GH¢)'], marker='x', linestyle='--', color='r', label='Predicted Prices')
plt.title('Average Price of Cement in Ghana (32.5R) 2009-2029')
plt.xlabel('Year')
plt.ylabel('Price (GH¢)')
plt.legend()
plt.grid(True)
plt.show()
```
